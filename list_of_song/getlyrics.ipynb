```
import csv
import numpy as np

# from google.colab import drive
# drive.mount('/content/gdrive')

with open("/content/gdrive/My Drive/test.csv", 'r') as csvfile:
    reader = csv.reader(csvfile)
    rows= [row for row in reader]
rows = rows[1:]
hot_list_100 = {}
for row in rows:
    hot_list_100[row[0]] = eval(row[1])
    print(row[0], len(eval(row[1])))

!pip install lyricsgenius
import lyricsgenius
# This is my lyricsgenius id
genius = lyricsgenius.Genius("lISrOoq8wD1T9e_Ta1ggsWuEnf9eeHfTWR-pSSV5puvg5ZYh-K9J9tX515Xg4wL8")

title = []
lyrics = []
artists = []
rank = []
cnt = []
date = []
# Try to get the lyrics and some other information of a song
# title, lyrics, artist, date
for year in hot_list_100:
    if year != '2001':
        continue
    hotlist = hot_list_100[year]
    # print(hotlist)
    for song in hotlist:
        try:
            s = genius.search_song(song)
            if s == None:
                continue
        except:
            continue
        else:
            title.append(song)
            lyrics.append(s.lyrics)
            artists.append(s.artist)
            date.append(year)
            rank.append(hotlist[song]['rank'])
            cnt.append(hotlist[song]['cnt'])
# there's a small part of song cannot be found, about 3%
import pandas as pd

dataframe = pd.DataFrame({'title':title, 'lyrics':lyrics, 'artists':artists, 'date':date, 'rank':rank, 'cnt': cnt})

dataframe.to_csv("/content/gdrive/My Drive/lyricsHotListTop_" + str(date[0]) + ".csv",index=True,sep=',')
print("success")
# Save the data in csv file
```
