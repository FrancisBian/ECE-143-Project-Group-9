```python
hot_during_20_years = []
# try to get top100 list from 1999 to 2019 
for i in range(1999, 2020):
    chart = billboard.ChartData('hot-100', str(i) + '-12-01')
    hot_during_20_years.append(chart)
    print(chart.date)
```

```python
print(hot_during_20_years)
```

```python
title = []
lyrics = []
artists = []
data = []
# Try to get the lyrics and some other information of a song
# title, lyrics, artist, date
for hotlist in hot_during_20_years:
    cnt = 0
    for song in hotlist:
        try:
            s = genius.search_song(song)
            if s == None:
                continue
            cnt += 1
        except:
            continue
        else:
            title.append(song)
            lyrics.append(s.lyrics)
            artists.append(s.artist)
            data.append(hotlist.date)
    print(hotlist.date, cnt)
# there's a small part of song cannot be found, about 3%
```
